```python
movie_name = 'whales_cut.mp4'
cap = cv2.VideoCapture(movie_name)
nframe =cap.get(cv2.CAP_PROP_FRAME_COUNT)
nx = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
ny = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
steps = 10
i = 0
while(cap.isOpened() & (i<=(nframe-steps))):
    i = i + steps
    cap.set(cv2.CAP_PROP_POS_FRAMES,i)
    ret, frame = cap.read()
    hls = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("title", hls)
    cv2.waitKey(0)

cap.release()
```

```python
movie_name = 'whales_cut.mp4'
cap = cv2.VideoCapture(movie_name)
nframe =cap.get(cv2.CAP_PROP_FRAME_COUNT)
nx = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
ny = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
steps = 10
i = 0
while(cap.isOpened() & (i<=(nframe-steps))):
    i = i + steps
    cap.set(cv2.CAP_PROP_POS_FRAMES,i)
    ret, frame = cap.read()
    ret,hls = cv2.threshold(frame,127,255,cv2.THRESH_BINARY)
    cv2.imshow("title", hls)
    cv2.waitKey(0)

cap.release()
```
